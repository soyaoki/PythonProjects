```python
import sys
import numpy as np
import matplotlib.pyplot as plt
from statistics import mean, median,variance,stdev
from scipy.stats import norm

import pandas as pd
import statsmodels.api as sm
```

```python
def read_txt(filename):
    data = []
    file = open(filename, 'r')

    for line in file.readlines():
        fname = line.rstrip().split(',') #using rstrip to remove the \n
        data.append(float(fname[1]))

    file.close()
    
    return data
```

```python
def ar_model(data):
    # モデルの生成
    model = sm.tsa.AR(data)
    # AICでモデルの次数を選択
    print(model.select_order(maxlag=6, ic='aic')) 

    # 推定
    result = model.fit(maxlag=6)

    # モデルが推定したパラメーター
    print(result.params)
    print(result.sigma2)

    # 予測
    res = model.predict(result.params,start=10, 
        end=len(data), 
        dynamic=False)

    # 表示
    plt.plot(data)
    plt.plot(res)
    plt.show()
    
    return result.params
```
