```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import glob
import pandas as pd
import collections
import numpy as np
import seaborn as sns

%matplotlib inline

plt.rcParams['font.family'] = "MS Gothic"

path = "./特実_国内文献.csv"
df = pd.read_csv(path)
df["出願年"] = pd.to_datetime(df["出願日"]).dt.year

app_top10_dic = {} #それぞれの年度のTOP10企業と出願数
app_total_dic = {} #それぞれの年度の全ての企業と出願数
app_all = [] #バブルチャートの対象となる企業のリスト

for year in sorted(df["出願年"].unique()):
    df_ = df[df["出願年"]==year]
    app_list = df_['出願人/権利者']
    app_list = [i.rstrip('他').replace(' ','').replace('\u3000',' ') for i in app_list]
    app_set = collections.Counter(app_list)
    app_top10 = app_set.most_common()[:10]
    app_all.extend([i[0] for i in app_top10])
    app_top10_dic[year] = app_top10

    app_total = app_set.most_common()
    app_total_dic[year] = app_total

app_all = list(set(app_all))
```

```python
years = list(app_top10_dic.keys())
df_timetable = pd.DataFrame(index=app_all,columns=years).fillna(0)
```

```python
for i in app_total_dic:
    dic = app_total_dic[i]
    for d in dic:
        if(d[0] in app_all):
            df_timetable[i][d[0]] += d[1]
```

```python
fig = plt.figure(figsize=(6,10),dpi=150)
sns.heatmap(df_timetable,square=True,cmap='Reds')
```

```python
fig = plt.figure(figsize=(6,10),dpi=150)

for n,y in enumerate(years):
    for m,x in enumerate(app_all[::-1]):
        plt.scatter(y,x,s=int(df_timetable[y][x])**2)

size = [1,5,10]
for s in size:
    plt.scatter(-100,-100,s=s**2,facecolor='white',edgecolor='black',label=str(s))
plt.legend(title='Patents', bbox_to_anchor=(1.03, 1), 
           loc='upper left',labelspacing=1,borderpad=1)

plt.xlim(2000,2025)
plt.ylim(-1,len(app_all))
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

# クレンジング

```python
import re
```

```python
# words_delete = ['<BR>',' ','\r','\n','【要約】','【選択図】','(57)']
words_delete = ['<BR>',' ','\r','\n','【要約】','【選択図】','(57)','ドライバ','車両','状態','推定','運転','制御','部','図','値','走行','システム','判定','装置','提供','方法','検出','算出']
```

```python
df["要約_オリジナル"] = df["要約"]
df["課題"] = ""
df["解決手段"] = ""

for i in range(len(df)):
    # 要約をきれいにする
    sentence = df["要約"].iloc[i]
    for word_del in words_delete:
        sentence = re.sub(word_del, '', sentence)
    df["要約"].iloc[i] = sentence
    
    # 課題と解決手段を抽出する
    kadai = re.findall(r'【課題】(.*)【解決手段】', sentence)
    kaiketsusyudan = re.findall(r'【解決手段】(.*)', sentence)
    if(len(kadai)!=0):
        df["課題"].iloc[i] = kadai[0]
    else:
        df["課題"].iloc[i] = "抽出不可"
    if(len(kaiketsusyudan)!=0):
        df["解決手段"].iloc[i] = kaiketsusyudan[0]
    else:
        df["解決手段"].iloc[i] = "抽出不可"
```

```python
# 形態素解析
import MeCab
m = MeCab.Tagger("-Ochasen")
def mecab_sep(text):
    node = m.parseToNode(text)

    words_list = []

    while node:
        if node.feature.split(",")[0] == "名詞":
            words_list.append(node.surface)
        elif node.feature.split(",")[0] == "動詞":
            words_list.append(node.feature.split(",")[6])
        elif node.feature.split(",")[0] == "形容詞":
            words_list.append(node.feature.split(",")[6])
            
        node = node.next
        
    return words_list[1:-1]

# pipで行ける分、楽
from janome.tokenizer import Tokenizer
def janome_sep(text):
    t = Tokenizer()
    tokens = t.tokenize(text)
    docs=[]
        
    for token in tokens:
        token_list = token.part_of_speech.split(",")
        if token_list[0] == "名詞" and token_list[1] != "非自立":
            docs.append(token.surface)
    return docs
```

```python
# ベクトル化
from sklearn.feature_extraction.text import TfidfVectorizer
from gensim.models.doc2vec import Doc2Vec

### Bag-of-words / tf-idf
def calc_vecs(docs):
    # vectorizer = TfidfVectorizer(analyzer=mecab_sep)
    vectorizer = TfidfVectorizer(analyzer=janome_sep)
    vecs = vectorizer.fit_transform(docs)
    return vecs.toarray()

### Doc2Vec
model = Doc2Vec.load("jawiki.doc2vec.dbow300d/jawiki.doc2vec.dbow300d.model")
def calc_vecs_d2v(docs):
    vecs = []
    for d in docs:
        # vecs.append(model.infer_vector(mecab_sep(d)))
        vecs.append(model.infer_vector(janome_sep(d)))
    return vecs
```

```python
from wordcloud import WordCloud
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

font_path="./ipaexg.ttf"

def show_wordcloud(series):
    long_string = ','.join(list(series.values))
    
    # Create a WordCloud object
    wordcloud = WordCloud(font_path=font_path, background_color="white", max_words=100, contour_width=3, contour_color='steelblue')
    
    # Generate a word cloud
    wordcloud.generate(long_string)
    
    # Visualize the word cloud
    plt.imshow(wordcloud)
    plt.show()
```

# 要約を分析する

```python
df["形態素解析_要約_mecab"] = " "
for i in range(len(df)):
    df["形態素解析_要約_mecab"].iloc[i] = mecab_sep(df["要約"].iloc[i])
    
df["形態素解析_要約_janome"] = " "
for i in range(len(df)):
    df["形態素解析_要約_janome"].iloc[i] = janome_sep(df["要約"].iloc[i])
```

```python
df[["形態素解析_要約_mecab","形態素解析_要約_janome"]]
```

```python
vec_yoyaku = calc_vecs_d2v(df["要約"])
# vec_yoyaku = calc_vecs(df["要約"])

df["Vector_yoyaku"] = vec_yoyaku
```

```python
import seaborn as sns
from scipy.cluster.hierarchy import linkage,dendrogram,fcluster
import matplotlib.pyplot as plt

linkage_result = linkage(vec_yoyaku, method = "ward", metric = "euclidean")

plt.figure(figsize=(50,20))
dendrogram(linkage_result, labels = df.index)
plt.show()
```

```python
from sklearn.cluster import KMeans

vec = KMeans(n_clusters=4)
vec.fit(vec_yoyaku)
```

```python
vec.labels_
```

```python
df["要約クラス_byVectorKMeans"] = vec.labels_
```

```python
clusters = fcluster(linkage_result, t=4, criterion='maxclust')
```

```python
clusters
```

```python
df["要約クラス_byVectorScipy"] = clusters
```
